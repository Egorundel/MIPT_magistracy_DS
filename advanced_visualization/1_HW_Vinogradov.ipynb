# Задание 1. Знакомство с данными и EDA

```python
# Импортируем библиотеки и данные
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

ny_housing = pd.read_csv("https://raw.githubusercontent.com/ElijahSum/mipt_visualization/master/week_01_visualization/data/NY-House-Dataset.csv")
ny_housing = ny_housing[ny_housing['PRICE'] <= 200000000]
ny_housing.head()
```

### Задание 1.1 Построить разные графики и сделать 6 выводов

```python
# График 1. Гистограмма цен на недвижимость
plt.figure(figsize=(14, 8))
sns.histplot(ny_housing['PRICE'], bins=50, kde=True)
plt.title('Распределение цен на недвижимость')
plt.xlabel('Цена')
plt.ylabel('Количество объектов')
plt.show()
```

```python
# График 2. Heatmap корреляции признаков
plt.figure(figsize=(14, 8))
sns.heatmap(ny_housing[['PRICE', 'BEDS', 'BATH', 'PROPERTYSQFT', 'LATITUDE', 'LONGITUDE']].corr(), annot=True, cmap='coolwarm')
plt.title('Тепловая карта корреляции признаков')
plt.show()
```

```python
# График 3. Scatter plot площади недвижимости и цены
plt.figure(figsize=(14, 8))
sns.scatterplot(x='PROPERTYSQFT', y='PRICE', data=ny_housing)
plt.title('Зависимость цены от площади недвижимости')
plt.xlabel('Площадь недвижимости')
plt.ylabel('Цена')
plt.show()
```

```python
# График 5. BarPlot средней цены по количеству ванных комнат 
plt.figure(figsize=(16, 8))
sns.barplot(x='BATH', y='PRICE', data=ny_housing.groupby('BATH')['PRICE'].mean().reset_index(), palette='coolwarm')
plt.title('Средняя цена в зависимости от количества ванных комнат')
plt.xlabel('Количество ванных комнат')
plt.ylabel('Средняя цена')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

### Задание 1.5 Какой тип квартир самый популярный в разных районах?

```python
# Наблюдения для каждого района
filtered_sublocalities = ny_housing['SUBLOCALITY'].value_counts()[ny_housing['SUBLOCALITY'].value_counts() >= 100].index
filtered_data = ny_housing[ny_housing['SUBLOCALITY'].isin(filtered_sublocalities)]
type_distribution = filtered_data.groupby(['SUBLOCALITY', 'TYPE']).size().unstack(fill_value=0)
type_distribution_pct = type_distribution.div(type_distribution.sum(axis=1), axis=0)
sorted_distribution_pct = type_distribution_pct.sort_values(by='Condo for sale', ascending=False)

sorted_distribution_pct.plot(kind='barh', stacked=True, figsize=(14, 10), colormap='coolwarm')
plt.title('Доля квартир разных видов по районам')
plt.xlabel('Доля')
plt.ylabel('Район')
plt.legend(title='Тип недвижимости')
plt.tight_layout()
plt.show()
```

### Задание 2.1 Поиск выбросов

В рамках рынка недвижимости, аномалии в переменных PRICE (цена) и PROPERTYSQFT (площадь имущества) могут быть обусловлены уникальными или необычными характеристиками объектов недвижимости, такими как чрезвычайно большие размеры или высокая стоимость из-за каких-либо особенностей или местоположения. В других числовых переменных, таких как количество спален (BEDS) или ванных комнат (BATH), также могут наблюдаться аномалии, но они менее склонны к большим колебаниям.

### Задание 2.2 Удаление выбросов

```python
# Расчет IQR
IQR = ny_housing['PRICE'].quantile(0.75) - ny_housing['PRICE'].quantile(0.25)

# Определение границ
lower_bound = ny_housing['PRICE'].quantile(0.25) - 1.5 * IQR
upper_bound = ny_housing['PRICE'].quantile(0.75) + 1.5 * IQR

# Построение графика
plt.figure(figsize=(14, 8))
barplot = sns.barplot(x='TYPE', y='PRICE', data=ny_housing[(ny_housing['PRICE'] >= lower_bound) & (ny_housing['PRICE'] <= upper_bound)], estimator=np.mean, errorbar=None, palette='coolwarm')
plt.xticks(rotation=45)
plt.title('Средняя стоимость недвижимости по типам после удаления выбросов')
plt.xlabel('Тип недвижимости')
plt.ylabel('Средняя стоимость')
plt.show()
```

### Задание 2.3 Поиск выбросов с помощью правила трёх сигм

Формулы для границ:  
$μ−3σ$  
$μ+3σ$

где μ - математическое ожидание, σ - стандартное отклонение

```python
# Математическое ожидание и стандартное отклонение:
math_expectation_sqft = ny_housing['PROPERTYSQFT'].mean()
std_sqft = ny_housing['PROPERTYSQFT'].std()

# Определение границ для экстремальных значений по формуле, приведенной выше (правило трёх сигм)
lower = math_expectation_sqft - 3*std_sqft
upper = math_expectation_sqft + 3*std_sqft

# Нахождение экстремальных значений (выбросов)
extreme_values = ny_housing[(ny_housing['PROPERTYSQFT'] < lower) | (ny_housing['PROPERTYSQFT'] > upper)]

print(f"Математическое ожидание: {math_expectation_sqft}")
print(f"Стандартное отклонение: {std_sqft}")
print(f"Количество экстремальных значений: {extreme_values.shape[0]}")
```

### Задание 2.4 Выделение выбросов

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='PROPERTYSQFT', y='PRICE', data=ny_housing[ny_housing['PROPERTYSQFT'] <= upper], color='#564EBF', label='Обычные значения')
sns.scatterplot(x='PROPERTYSQFT', y='PRICE', data=ny_housing[ny_housing['PROPERTYSQFT'] > upper], color='#FFCC00', label='Выбросы')
plt.title('Распределение цен в зависимости от площади с выбросами')
plt.xlabel('Площадь собственности')
plt.ylabel('Цена')
plt.legend()
plt.show()

filtered_ny_housing = ny_housing[ny_housing['PROPERTYSQFT'] <= upper]

print("Количество строк до удаления:", ny_housing['PROPERTYSQFT'].shape[0])
print(f"Количество строк, подлежащих удалению: {ny_housing['PROPERTYSQFT'].shape[0] - filtered_ny_housing.shape[0]}")
print("Количество строк после удаления:", filtered_ny_housing.shape[0])
```

# Задание 3. Создание простой модели, визуализация результатов работы алгоритма

### Задание 3.1 Создание модели парной регрессии

```python
# Импорт необходимых библиотек
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Используем фильтеред из предыдущего задания
X = filtered_ny_housing[['PROPERTYSQFT']].values  
y = filtered_ny_housing['PRICE'].values  

# Создание и обучение модели линейной регрессии
model = LinearRegression()
model.fit(X, y)

# Вывод коэффициентов модели
print(f"Коэффициент (slope): {model.coef_[0]}")
print(f"Перехват (intercept): {model.intercept_}")
```

### Задание 3.2 Визуализация результата

```python
plt.figure(figsize=(14, 8))
plt.scatter(X, y, color='#564EBF', label='Наблюдаемые значения')
plt.plot(X, model.predict(X), color='#FFCC00', label='Прямая регрессионного уравнения')
plt.title('Парная регрессии цены и площади собственности')
plt.xlabel('Площадь собственности')
plt.ylabel('Цена')
plt.legend()
plt.show()
```
